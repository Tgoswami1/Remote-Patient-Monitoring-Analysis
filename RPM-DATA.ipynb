import pandas as pd
import numpy as np

# Number of users
num_users = 100

# Create mock data
data = {
    "User ID": range(1, num_users + 1),
    "Setup Completed": np.random.choice([0, 1], num_users),  # 0 = No, 1 = Yes
    "Data Syncs": np.random.randint(0, 20, num_users),  # Random number of data syncs
    "Retention Period (Months)": np.random.choice(range(1, 13), num_users),  # Retention in months
    "Health Alerts Triggered": np.random.randint(0, 5, num_users),  # Random health alerts
}

df = pd.DataFrame(data)

# Save the dataset to a CSV file
df.to_csv("mock_rpm_data.csv", index=False)

print("Mock dataset created.")


```python
import pandas as pd
import numpy as np
# Loading dataset
df = pd.read_csv("mock_rpm_data.csv")

# Calculating metrics
setup_completion_rate = df["Setup Completed"].mean() * 100  # in percentage
average_syncs = df["Data Syncs"].mean()
average_retention = df["Retention Period (Months)"].mean()
average_alerts = df["Health Alerts Triggered"].mean()

# Print metrics
print(f"Setup Completion Rate: {setup_completion_rate:.2f}%")
print(f"Average Data Sync Frequency: {average_syncs:.2f} syncs per user")
print(f"Average Retention Period: {average_retention:.2f} months")
print(f"Average Health Alerts Triggered: {average_alerts:.2f} alerts per user")
```

### Setup Completion Rate:

```python
setup_completion_rate = df['Setup Completed'].mean() * 100
print(f"Setup Completion Rate: {setup_completion_rate:.2f}%")
```

### Average Data Sync Frequency:

```python
avg_data_sync = df['Data Syncs'].mean()
print(f"Average Data Sync Frequency: {avg_data_sync:.2f} syncs per user")
```

## Average Retention Period:

```python
avg_retention_period = df['Retention Period (Months)'].mean()
print(f"Average Retention Period: {avg_retention_period:.2f} months")
```

### Average Health Alerts Triggered:

```python
avg_health_alerts = df['Health Alerts Triggered'].mean()
print(f"Average Health Alerts Triggered: {avg_health_alerts:.2f} alerts per user")
```

### Bar Graph for Setup Completion

```python
import matplotlib.pyplot as plt
df['Setup Completed'].value_counts().plot(kind='bar')
plt.title('Setup Completion Distribution')
plt.xlabel('Setup Completed (1 = Yes, 0 = No)')
plt.ylabel('Number of Users')
plt.show()
```

```python

import seaborn as sns

# Bar chart for metrics
metrics = ['Setup Completion Rate', 'Average Syncs', 'Average Retention', 'Average Alerts']
values = [setup_completion_rate, average_syncs, average_retention, average_alerts]

plt.figure(figsize=(8, 6))
sns.barplot(x=metrics, y=values)
plt.title('RPM Metrics Comparison')
plt.ylabel('Values')
plt.show()
```
